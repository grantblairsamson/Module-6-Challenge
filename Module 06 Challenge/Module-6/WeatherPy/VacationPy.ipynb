```python
# Import necessary libraries
import pandas as pd
import requests
import geoviews as gv
import geoviews.feature as gf
from geoviews import opts
import hvplot.pandas
from api_keys import geoapify_key  # Correctly import the Geoapify API key
```

```python
# Load the weather data generated from WeatherPy
city_data_df = pd.read_csv(r'C:\Users\User\OneDrive\Desktop\(Copy) bootcamp\Github challenges or Projects\Module 06 Challenge\Module-6\output_data\city_data.csv')
```

```python
# Step 1: Display Cities on a Map
city_data_df.hvplot.points("Lng", "Lat", geo=True, tiles="OSM", size="Humidity", color="Humidity",
                           alpha=0.5, frame_width=700, frame_height=500, hover_cols=["City", "Country"])
```

```python
# Step 2: Filter for Ideal Vacation Conditions
ideal_conditions = city_data_df[
    (city_data_df["Max Temp"] > 21) & (city_data_df["Max Temp"] < 27) &
    (city_data_df["Wind Speed"] < 4.5) &
    (city_data_df["Cloudiness"] == 0)
]

ideal_conditions.head()
```

```python
# Step 3: Find Nearby Hotels Using Geoapify API
# Create a new DataFrame to store hotel information
hotel_df = pd.DataFrame(columns=["City", "Country", "Lat", "Lng", "Hotel Name", "Humidity"])

# Loop through the filtered cities and find the first hotel within 10,000 meters
for index, row in ideal_conditions.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    
    # API request to Geoapify
    query_url = f"https://api.geoapify.com/v2/places?categories=accommodation.hotel&filter=circle:{lng},{lat},10000&limit=1&apiKey={geoapify_key}"
    response = requests.get(query_url).json()
    
    # Extract hotel information
    if response["features"]:
        hotel_name = response["features"][0]["properties"]["name"]
        
        # Create a DataFrame with the new hotel data
        new_hotel = pd.DataFrame({
            "City": [row["City"]],
            "Country": [row["Country"]],
            "Lat": [lat],
            "Lng": [lng],
            "Hotel Name": [hotel_name],
            "Humidity": [row["Humidity"]]
        })
        
        # Concatenate only if the new_hotel DataFrame is not empty and does not have all NA values
        if not new_hotel.isna().all().all():
            hotel_df = pd.concat([hotel_df, new_hotel], ignore_index=True)

# Display the resulting DataFrame with hotel information
hotel_df.head()
```
